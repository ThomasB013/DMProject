```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data_points.txt", skiprows=1, dtype=float)
x = np.array([row[0] for row in data])
y = np.array([row[1] for row in data])

plt.scatter(x, y)
plt.title("Plot of data_points.txt with outliers. The data model is y = 1/35x^3 - 15x + 10.")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
data_outliers = np.loadtxt("outlier_correct.txt")

plt.scatter(x, y, c=data_outliers)
plt.title("Outlier detection on data_points.txt each colour represents an iteration.")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(x[data_outliers == 0], y[data_outliers ==0])
plt.title("data_points.txt with the outliers removed.")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
data_outliers = np.loadtxt("outlier_incorrect.txt")

plt.scatter(x, y, c=data_outliers)
plt.title("Wrong outlier detection on data_points.txt each colour represents an iteration. ")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(x[data_outliers == 0], y[data_outliers ==0])
plt.title("data_points.txt with the incorrect outliers removed.")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
